```python
import matplotlib.pyplot as plt

from adaboost import AdaBoost
from utils import make_moons
```

```python
n_samples = 100
noise = 0.25

X, y = make_moons(n_samples = n_samples, noise = noise)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='viridis')
plt.title("data")
plt.show()
```

```python
# ------------------------------------------
# Parameters for the AdaBoost algorithm
n_classifiers = 30
plotting = False
create_gif = True
# ------------------------------------------

ada = AdaBoost(n_classifiers = n_classifiers)
ada.initialize_adaboost(X, y)


for _ in range(n_classifiers):
    
    # decision step
    ada.fit_step()

    ada.plot(plotting = plotting)


if create_gif:
    from utils import gif
    gif(duration = 500)
```
